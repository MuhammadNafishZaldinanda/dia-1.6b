# Indspeech-news-tts

## Audio Length

```python
import os
import torch
import torchaudio
import dac
from tqdm import tqdm
import pandas as pd


import os
os.environ["CUDA_VISIBLE_DEVICES"] = "2"

wav_dir = "/mnt/ssd/nafish/text-to-speech/data/indspeech_news_tts/wavs"
device = "cuda" if torch.cuda.is_available() else "cpu"

dac_model = dac.DAC.load(dac.utils.download()).to(device)

wav_files = [os.path.join(wav_dir, f) for f in os.listdir(wav_dir) if f.lower().endswith(".wav")]

lengths = []
for path in tqdm(wav_files, desc="Processing audio"):
    try:
        wav, sr = torchaudio.load(path)
        wav = wav.to(device)
        with torch.inference_mode():
            result = dac_model.encode(wav.unsqueeze(0))
            codes = result[0] if isinstance(result, tuple) else result
            T = codes.shape[-1]
            lengths.append(T)
    except Exception as e:
        print(f"❌ Error on {path}: {e}")

series = pd.Series(lengths)
print("\n=== Audio Frame Stats (DAC tokens) ===")
print(f"Total files : {len(lengths)}")
print(f"Max         : {series.max()}")
print(f"95th %ile   : {int(series.quantile(0.95))}")
print(f"Mean        : {int(series.mean())}")
print(f"Min         : {series.min()}")
```

```python
import pandas as pd
from g2p_id import G2P

# Inisialisasi G2P
g2p = G2P()

# Path ke file CSV input dan output
input_csv = "/mnt/ssd/nafish/text-to-speech/data/combined_dataset.csv"  # ganti sesuai file kamu
output_csv = "/mnt/ssd/nafish/text-to-speech/data/combined_dataset_phoneme.csv"

# Baca CSV
df = pd.read_csv(input_csv, sep="|", header=None, names=["audio_path", "grapheme_text"])

# Ubah grapheme ke phoneme
df["phoneme_text"] = df["grapheme_text"].apply(g2p)

# Tulis ulang CSV hanya dengan audio_path dan phoneme_text
df[["audio_path", "phoneme_text"]].to_csv(output_csv, sep="|", index=False, header=False)
```

```python
import pandas as pd

phoneme_csv_path = "/mnt/ssd/nafish/text-to-speech/data/combined_dataset_phonem.csv"

# Baca file CSV hasil phonemization
df = pd.read_csv(phoneme_csv_path, sep="|", header=None, names=["audio_id", "phoneme_text"])

# Hitung panjang byte dari phoneme_text
byte_lengths = df["phoneme_text"].apply(lambda x: len(str(x).encode("utf-8")))
series = byte_lengths

print("\n=== Phoneme Text Byte Stats (Dia-1.6B style) ===")
print(f"Total samples: {len(series)}")
print(f"Max bytes    : {series.max()}")
print(f"95th %ile    : {int(series.quantile(0.95))}")
print(f"Mean bytes   : {int(series.mean())}")
print(f"Min bytes    : {series.min()}")
```
