# Voice Clone

```python
from dia.model import Dia


model = Dia.from_local(
    config_path= "../config_inference.json",
    checkpoint_path="../ckpt.pth",
    device="cuda",)

clone_from_text = "Hari ini saya tidak pergi ke sekolah karena demam."
clone_from_audio = "clone.mp3"

text_to_generate = "Dikarenakan kasus covid semakin tinggi, status PPKM di DIY ditingkatkan ke level empat."

# It will only return the audio from the text_to_generate
output = model.generate(
    clone_from_text + text_to_generate, 
    audio_prompt=clone_from_audio, 
    use_torch_compile=True, 
    verbose=True, 
    cfg_scale=3.0,
    temperature=1.3,
    top_p=0.95,
    cfg_filter_top_k=35,
)

sf.write("voice_clone.mp3", output, 44100)
```
