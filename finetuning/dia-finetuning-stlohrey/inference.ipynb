# Voice Clone with Speed Factor

```python
import soundfile as sf
from dia.model import Dia
import numpy as np

# Load model
model = Dia.from_local(
    config_path="config.json",
    checkpoint_path="ckpt_epoch1.pth",
    device="cuda",
)

# Audio dan teks referensi (untuk voice cloning)
clone_from_text = "Mural kota Semarang. Sejarah kota Semarang dimulai sejak abad ke delapan masehi."
clone_from_audio = "tasya-f-indonesia-naration-formal.mp3"

# Teks yang ingin digenerate
text_to_generate = "Halo apa kabar ?"

# Generate audio (output = waveform)
output_waveform = model.generate(
    clone_from_text + text_to_generate, 
    audio_prompt=clone_from_audio, 
    use_torch_compile=True, 
    verbose=True, 
    cfg_scale=3.0,
    temperature=1.3,
    top_p=0.95,
    cfg_filter_top_k=35,
)

# Asumsikan sample rate output (standarnya: 24000 Hz)
output_sr = 44100

# Ubah kecepatan (jika diperlukan)
speed_factor = 1.5   # Ubah ke 0.8, 1.2, dll jika ingin lambat/cepat
if speed_factor != 1.0:
    original_len = len(output_waveform)
    target_len = int(original_len / speed_factor)
    x_original = np.arange(original_len)
    x_resampled = np.linspace(0, original_len - 1, target_len)
    resampled_audio_np = np.interp(x_resampled, x_original, output_waveform)
    output_waveform = resampled_audio_np.astype(np.float32)
    print(f"Audio disesuaikan dari {original_len} ke {target_len} sampel (speed {speed_factor:.2f}x)")

# Simpan hasil audio ke file
sf.write("output_dia_tts.wav", output_waveform, output_sr)
```
