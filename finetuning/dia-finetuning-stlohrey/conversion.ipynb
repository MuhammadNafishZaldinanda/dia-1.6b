```python
import requests

clone = "Dari waktu ke waktu inavoice telah bekerja sama dengan berbagai brand maupun perusahaan perusahaan ternama."
text = "Selamat sore Bapak Ibu sekalian, selamat datang di acara pembukaan rapat tahunan dua ribu dua puluh dua."
# text = "Acara terakhir adalah ramah tamah, silakan diambil kudapannya di samping meja."
# text = "Dikarenakan kasus covid semakin tinggi, status pe pe ka em di de i ye ditingkatkan ke level empat."

headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json',
}

json_data = {
    'model': '/mnt/ssd/nafish/text-to-speech/dia-finetuning/ft-combined-dataset-22072025',
    'input': clone + text,
    'voice': 'English',
    'response_format': 'mp3',
    'speed': 0.25,
}

response = requests.post('http://172.16.245.32:9020/v1/audio/speech', headers=headers, json=json_data)

if response.status_code == 200:
    with open("vox-box-predict-0.25.mp3", "wb") as f:
        f.write(response.content)
    print("Audio berhasil disimpan")
else:
    print("Gagal:", response.status_code, response.text)
```
